```python
import pandas as pd 
from datetime import datetime
import numpy as np 
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
Standard_Poors = pd.read_csv("C:/Users/Nicolas/Documents/sphist.csv")
Standard_Poors[0:10]
Standard_Poors["Date"] = pd.to_datetime(Standard_Poors["Date"])
```

```python
#Sort the Standard_Poors dataframe on the Date column
Standard_Poors = Standard_Poors.sort_values(by="Date")
Standard_Poors.head(10)
```

I will compute some indicators over the Standard_Poors data frame that will be the features of the model, then  I will use these indicators with the target variable to fit the model. These indicators are Average price from the past 5 days, average price from the past 30 days, standard deviation of the price over the past 365 days, average volume for the past 5 days,  standard deviation of the volume for the past 5 days. 


 In all of these indicators it is necessary to use the method shift due to not include the current row in the 
value of the indicator. The model must learn how to predict the current price with historical prices. 


```python
# 5 days average close price.   
average_5 = Standard_Poors["Close"].rolling(center=False,window=5).mean()
Standard_Poors.head(10)
average_5 = average_5.shift(periods=1)
Standard_Poors["average_5"] = average_5 
Standard_Poors[0:10]
```

```python
# 30 days average close price.
average_30 = Standard_Poors["Close"].rolling(center=False,window=30).mean()
average_30 = average_30.shift(periods=1)
Standard_Poors["average_30"] = average_30
Standard_Poors[0:10]
```

```python
# standard deviation of the price over 365 last days. 
std_365 = Standard_Poors["Close"].rolling(center=False,window= 365).std()
std_365 = std_365.shift(periods=1)
Standard_Poors["std_365"] = std_365 
Standard_Poors
```

```python
# average volume of the past 5 days
average_volume_5 = Standard_Poors["Volume"].rolling(center=False,window=5).mean()
average_volume_5 = average_volume_5.shift(periods=1)
Standard_Poors["average_volume_5"] = average_volume_5
```

```python
# standard deviation of the volume of the past 5 days 
std_average_volume_5 = Standard_Poors["Volume"].rolling(center=False,window=5).std()
std_average_volume_5 = std_average_volume_5.shift(periods=1)
Standard_Poors["std_average_volume_5"] = std_average_volume_5 
Standard_Poors.tail()
```

```python
# The indicators above are the features of the model. Some of these indicators need 1 year of prices to compute. 
# For this reason is necessary to remove rows before 1951-01-03 because these rows contain missing values fore these 
# indicators.  
Standard_Poors = Standard_Poors[Standard_Poors["Date"] > datetime(year=1951, month=1, day=2)]

# To train the model is necessary to drop any missing value in the dataset.
Standard_Poors = Standard_Poors.dropna(axis=0)

# Generate the train and test dataframes to make predictions of the close values of the Standard&Poors index.
train = Standard_Poors[Standard_Poors["Date"] < datetime(year=2013, month=1, day=1)]
test = Standard_Poors[Standard_Poors["Date"] >= datetime(year=2013, month=1, day=1)]
test
```

```python
# In this step, I will fit the model with the LinearRegression classifier. We are trying to predict the close value 
# of the Standard and Poors index. So the target of the model is the "Close" Column and the features that I will use
# in this step are the average_5, average_30, std_365 that are already calculated above.
lr = LinearRegression()
X_train = train[["average_5","average_30","std_365"]]
Y_train = train["Close"]

lr.fit(X_train,Y_train)                   
```

```python
# Make Predictions using the test dataset and the features average_5, average_30 and std_365

X_test = test[["average_5","average_30","std_365"]] # The test dataframe with the predictors, that will use to predict the target value
Y_test = test["Close"]                              # The labels of the model

close_predictions = lr.predict(X_test)              # Will predict the close values of the index in the test dataframe

print(close_predictions[0:5])
print(Y_test[0:5])
```

```python
#Compute the Mean Absolute Error of the model
mae = sum(abs(close_predictions - test["Close"].values)) / test.shape[0]
print(len(close_predictions))
print(mae)
```

The model yield a Mean Absolute Error of 16.1751. The MAE is the mean absolute value of the differences between the predictions and the real value of the target variable. This metric is more intuitive than others such as the Mean Squared Error, in terms of how close the predictions were to the real price. 

```python
# In this step, I will add 2 more features to the model, and observe if the error metric can be reduced. These features are the 
# average 5 days volume (average_volume_5) and the standard deviation of the 5 days volume(std_average_volume_5) of the S&P index.
# Volume is the amount of trading contracts(buy and sell) of the index in a given day. The average volume of the Standard and Poors
# could be important to identify accumulations and distributions days of the index that can be used to get a sense of the momentum of the index in a certain time 
# and predict future movements.

X_train = train[["average_5","std_365","average_30","average_volume_5","std_average_volume_5"]]
Y_train = train["Close"]
lr.fit(X_train,Y_train)
```

```python
X_test = test[["average_5","std_365","average_30","average_volume_5","std_average_volume_5"]]

close_predictions = lr.predict(X_test)
print(close_predictions[0:5])
print(test["Close"][0:5])
```

```python
#Compute the Mean Absolute Error including the new features in the model
mae = sum(abs(close_predictions - test["Close"].values)) / test.shape[0]
print(mae)
```

With the two new features that were incorporated to the model the error of the model go up some amount, from 16.17 to 16.24. This imply that the new features doesn´t improve the performance of the model, and it will be necessary to do some research in the features side in order to minimize the error metric. 

```python
# In this step, I will predict the price of the Standard & Poors Index only one day ahead instead along the entire test
# dataframe.The approach that I will take is to start with the train dataset up to the date 2013-01-03 and within a 
# for loop with the length of the test dataframe, increment the train dataset by one day, create a new test dataframe
# that take one row of the  Standard_Poors with the current date, fit a new model and making a prediction for the 
# following day
import numpy as np
import datetime
from datetime import timedelta
from sklearn.metrics import mean_absolute_error

train = Standard_Poors[Standard_Poors["Date"] < datetime.datetime(year=2013, month=1, day=3)]
test_one_day = Standard_Poors[Standard_Poors["Date"] == datetime.datetime(year=2013, month=1, day=3)]



lr.fit(train[["average_5","average_30","std_365","average_volume_5","std_average_volume_5"]],train["Close"])   
prediction = lr.predict(test_one_day[["average_5","average_30","std_365","average_volume_5","std_average_volume_5"]])

date = datetime.datetime(year= 2013, month= 1, day=2)
mae = []
for i in range(0,len(test)):
    date += datetime.timedelta(days=1)
    train = Standard_Poors[Standard_Poors["Date"] < date]
    try:
        test_one_day = Standard_Poors[Standard_Poors["Date"] == date]
        lr.fit(train[["average_5","average_30","std_365","average_volume_5","std_average_volume_5"]],train["Close"])
        prediction = lr.predict(test_one_day[["average_5","average_30","std_365","average_volume_5","std_average_volume_5"]])                    
        mae.append(abs(prediction - test_one_day["Close"].values))

    except:
        pass

mae = np.mean(mae)
print(mae) 
```

When I use a dynamic train set of features and predict the next day over the test set, the mae fall to 14.5947. This error metric is sensible less than the one that I get when the predictions was for the total test set. 

```python
# In this step, I would predict the direction of the Standard_Poor index using the Random Forest Classifier. First, I downloaded the data and 
# read as a data frame in Pandas to start generate the features.

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

Standard_Poors = pd.read_csv("C:/Users/Nicolas/Documents/sphist.csv")
Standard_Poors["Date"] = pd.to_datetime(Standard_Poors["Date"])

#Sort the Standard_Poors dataframe on the Date column
Standard_Poors = Standard_Poors.sort_values(by="Date")

```

```python
# The next step I would try is to use the Random Forest Classifier and evaluate it performance. In this case, 
# I will choose the lagged percent changes of the prices as predictable indicators or features
# Create the shifted lag series of prior trading period close values 

for i in range(0, 5): 
    Standard_Poors["Lag%s" % str(i+1)] = Standard_Poors["Adj Close"].shift(i+1)
Standard_Poors[0:10]
```

```python
# Create a column with the returns of the Standard and Poors
Standard_Poors["return"] = Standard_Poors["Close"].pct_change()*100
```

```python
# Replace all the cero values in the return column of the Standard_Poors DataFrame with 0.0001 in order to allow make calculations with the return column
Standard_Poors.loc[Standard_Poors["return"] == 0,"return"] = 0.0001
      
```

```python
# Create columns with the lagged returns of the Standard_Poors dataframe
for i in range(0, 5): 
    Standard_Poors["lagged_return%s" % str(i+1)] =  Standard_Poors["Lag%s" % str(i+1)].pct_change()*100.0
#Standard_Poors[["Adj Close","return","lagged_return1","lagged_return2","lagged_return3","lagged_return4","lagged_return5"]]
Standard_Poors[0:10]
```

```python
# Create the column direction that will be the target variable of the model and can take a 
# value of 1 or -1. If the percentage change of the return is positive, the value of column Direction will be 1, and if the percentage
# change is negative, the value of the column Direction will be -1.
Standard_Poors["Direction"] = np.sign(Standard_Poors["return"])
Standard_Poors[0:10]
```

```python
# Generate the train and test data frames with the new features and target. The predictors of the models will be 
# the lagged_return1 and lagged_return2 indicators of the percentage change of the price.
from datetime import datetime
Standard_Poors = Standard_Poors[Standard_Poors["Date"] > datetime(year=1951, month=1, day=2)]

# Drop any missing value to fit the model
Standard_Poors = Standard_Poors.dropna(axis=0)

train_rf = Standard_Poors[Standard_Poors["Date"] < datetime(year=2013, month=1, day=1)]
test_rf = Standard_Poors[Standard_Poors["Date"] >= datetime(year=2013, month=1, day=1)]

X_train = train_rf[["lagged_return1","lagged_return2"]]
y_train = train_rf["Direction"]
X_test = test_rf[["lagged_return1","lagged_return2"]]
y_test = test_rf["Direction"]
train_rf.head()
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
print(predictions[0:20])
accuracy = clf.score(X_test,y_test)
accuracy
```

```python
# Asses the accuracy with the confussion matrix metric that the sklearn library provide
confusion_matrix(predictions,y_test)
```

The confusion matrix show that the RandomForestClasiffier has more probability of correct predict when the market goes down than when the market go up. The matrix show that the algorithm predict 190 times the correct direction of the market when the market go down while 168 times predict incorreclty classified down periods.

In the other hand when the market go up, 169 times the classifier predict correct up periods, while 212 the classifier incorrectly predict up periods(predict that the market go down when the market go up).
