```python
# Figure out how many people eat Pies, and what Pies people eat
apple_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pumpkin_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
pecan_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])

ate_pies = (apple_isnull) & (pumpkin_isnull) & (pecan_isnull)
ate_pies.value_counts()
```

```python
# In this step, I extract the first item of the column Age. Example: column Age = 18 - 29, the value will be 18. 
# Remove the + character from 60+ and convert the column Age to numeric(int)
def convert_age(row):
    age = row['Age']
    age_is_null = pd.isnull(age)
    if age_is_null == True:
        return None
    age_split = age.split( )[0]
    age_remove = age_split.rstrip("+")
    age_int = int(age_remove)
    return age_int
data['int_age'] = data.apply(convert_age,axis=1)


    
```

```python
data['int_age'].describe()
```

This metodology has the disadvantage that take the first number of the interval. If the age of the respondent are between 30 - 44, the age will be 30, and if the respondent is older than 60, the function assign a age of 60. 

I think that is important to take into account the withdraw of this method because assign people less age than they have. 

```python
def convert_income(row):
    income = row['How much total combined money did all members of your HOUSEHOLD earn last year?']
    income_null = pd.isnull(income)
    if income_null == True:
        return None
    income_split = income.split( )[0]
    if income_split == 'Prefer':
        return None
    income_clean = income_split.replace("$","")
    income_clean_all = income_clean.replace(",","")
    income_int = int(income_clean_all)
    return income_int
data['int_income'] = data.apply(convert_income,axis=1)
```

```python
data['int_income'].describe()
```

The results of the int_income function, take the first item of the category in the value of the column 'How much total combined money did all members of your HOUSEHOLD earn last year?'. 
This could introduce a downward bias in the result, because the methodology assign to the respondent, the lower value of the range category. The people that are within the first range($0 to $9,999) will have an income of 0.
