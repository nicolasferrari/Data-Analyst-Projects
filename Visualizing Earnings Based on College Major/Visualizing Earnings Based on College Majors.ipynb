```python
# Analyze the distribution of median salaries and unemployment rate
fig,ax = plt.subplots()
ax.boxplot(recent_grads["Median"])
ax.set_ylim(0,80000)
ax.set_xticklabels(["Salary"])
```

In the median boxplot, is possible to observe that the first quartile of the distribution(0-25%) is within the range 21000-31000, while the 2nd quartile is the shortest with a bounded interval between 32000-35000. 
The third quartil goes between 35000 to 45000 and the last queartile is the largest with an interval of 45000-60000.


```python
#Analyze the distribution of the unemployment rate variable
fig,ax = plt.subplots()
ax.boxplot(recent_grads["Unemployment_rate"])
#ax.set_ylim(0,80000)
ax.set_xticklabels(["Unemployment rate"])
```

The boxplot of the unemployment rate show that the median of the median of the unemployment rate is around 7% and the range of the variable are between 0 to 13%.
