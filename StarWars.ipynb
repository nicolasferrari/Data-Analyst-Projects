```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv",encoding= "ISO-8859-1")

# Exploring the data
star_wars.head(10)
star_wars.columns

# Remove rows from the Data Frame where there are NaN values on
# the RespondentID column

star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]

```

```python
# Use a dictionary to define a mapping from each value to a new value
# Map the variables Have you seen any of the 6 films in the Star Wars franchise ?
# Map the variable Do you consider yourself to be a fan of the Star Wars film franchise? 

yes_no = {"Yes": True, "No": False}

# First variable 

for col in [
    'Have you seen any of the 6 films in the Star Wars franchise?',
    'Do you consider yourself to be a fan of the Star Wars film franchise?'
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
import numpy as np
# Convert into a boolean values the columns 3 to 8 of the star_wars data frame
mapping_dict = {
    'Star Wars: Episode I  The Phantom Menace': True, 
    np.nan: False,
    'Star Wars: Episode II  Attack of the Clones': True, 
    'Star Wars: Episode III  Revenge of the Sith': True, 
    'Star Wars: Episode IV  A New Hope': True, 
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True}
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(mapping_dict)


# Rename the columns to better reflect what the values represent 
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", 
        "Unnamed: 4": "seen_2", 
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5", 
        "Unnamed: 8": "seen_6"})
star_wars.head(39)
```

```python
print(star_wars[star_wars.columns[3:9]])
```

```python
# The columns 9 to 14 of the Data Frame ask the respondent to rank the Star Wars movies in order of least to most favorite. Each column is the movie name, for example:
# Unnamed:10 means How much the respondent liked the movie Star Wars: Episode II Attack of the Clones
# First I convert the values of these columns to float numbers
# Then I rename the columns to clarify their sense

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4", 
        "Unnamed: 13": "ranking_5", 
        "Unnamed: 14": "ranking_6"})
```

```python
# In this step I will find the highest ranked movie
# For this purpose I calculate the mean of each ranking movie
star_wars[star_wars.columns[9:15]].mean()
```

```python
# Make a bar plot to observe the mean ranking of each movie
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(6)
labels = ['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']
plt.xticks(x,labels)
plt.bar(x,star_wars[star_wars.columns[9:15]].mean())

plt.show()
```

To make the bar plot, firstly I convert the values of the columns 9 to 14 to float types to enable to do arithmetic operations. Then I rename the variables to have a more descriptive name. These names are ranking_1,ranking_2,..., ranking_6. 

Lastly I calculate the mean of each column to find out which movie has the best ranking. The criteria to rank the movie vary across 1 to 6 with the value of 1 been the best ranking value and the value of 6 the worst ranking value.

With this critera, the movie that has the less mean will have the best ranking. 

The bar plot show that the movie placed in the column ranking_5 has the best rating within the 6 movies. This movie is 'Star Wars: Episode V The Empire Strikes Back'.

```python
# Find the most seen movie: calculate the sum to the columns numbert 3 to 9. These columns
# have booleans values. When call the methods like sum() or mean(), booleans are treated like Integers. True is treated
# like a 1 and False is treated like a 0. 
star_wars[star_wars.columns[3:9]].sum()
```

```python
# Make a bar chart of the rankings in columns 3 to 9.
x = np.arange(6)
labels = ['seen_1','seen_2','seen_3','seen_4','seen_5','seen_6']
plt.xticks(x,labels)
plt.bar(x, star_wars[star_wars.columns[3:9]].sum())
plt.show()
```

This bar plot show the ranking of most seen movie. Each column(seen_1,seen_2,seen_3,...,seen_6) is a movie, and the height of each bar is the sum of people that have seen the movie. Therefore the most seen movie is that in column 'seen_5' that is 'Star Wars: Episode V The Empire Strikes Back'. 

This result make sense because that movie has also the best ranking. This follows of the last bar plot. 

```python
# In this step I split the Data Frame in two parts based on the gender column.
# This kind of segmentation  allow to compute calculations for the 2 subsets of the data frame
# and observe differences


males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

males_mean = males[males.columns[9:15]].mean()
females_mean = females[females.columns[9:15]].mean()

print('Ranking for males:\n',males_mean)
print('Ranking for females:\n',females_mean)
```

```python
males_sum = males[males.columns[3:9]].sum()
females_sum = females[females.columns[3:9]].sum()
print('Most seen by males:\n', males_sum)
print('Most seen by females:\n', females_sum)
```

When the Data Frame is split between gender variables, the main result doesn't change. This imply that the movie 'Star Wars: Episode V The Empire Strikes Back' is the most seen and has the best rating by the two genders.

```python
# It will be interesting to split the Data Frame by the Education of the respondents
# First, observe the values that this variable take
star_wars['Education'].value_counts()
```

```python
split_Education = star_wars.groupby('Education').mean()
split_Education[split_Education.columns[8:15]]
```

With the Education segmentation, the result shows that the movie with the best ranking is in column ranking_5 which is the Star Wars: Episode V The Empire Strikes Back movie. 

The worst ranking for almost all the categories is the movie allocated in ranking_3 that is the Star Wars: Episode III Revenge of the Sith movie.

The unique difference in relation of the rest of the result is that in the category of 'Less than high school degree', the movie with the less ranking is Star Wars: Episode I The Phantom Menace, allocated in column ranking_1.  

```python
split_Education = star_wars.groupby('Education').sum()
split_Education[split_Education.columns[2:8]]
```

Based on the education of the respondents, the most seen movie for all the categories is the movie en the seen_5 column. This conclusion is the same, if we split the survey into gender variables. 

In the other hand, the movie with the least viewers is Star Wars: Episode III revenge of the Sith allocated in the seen_3 column. This conclusion is also the same, of that we have arrived in by the gender segmentation of the sample. 

```python
# Next, I conduct the same analysis but with the segmentation of the sample by the Location of the respondents
star_wars["Location (Census Region)"].value_counts()
        
split_location = star_wars.groupby('Location (Census Region)').mean()
split_location[split_location.columns[8:15]]


```

When the sample is split by Location, the conclusion doesn't change because the movie with the best ranking is the Star Wars: Episode V The Empire Strikes Back allocated in ranking_5 column. Likewise, the movie with the worst ranking is Star Wars: Episode III Revenge of the Sith, allocated in ranking_3 column. This is the same that have the lowest ranking when the sample is splited by gender or education. 

```python
# Examine which movie by Location, has seen most by the people in each location.
split_location = star_wars.groupby('Location (Census Region)').sum()
split_location[split_location.columns[2:8]]
```

Analyzing the result of how many people have seen each movie, the result are the same that when the sample was spliting by Education or Location. Movie Star Wars: Episode V The Empire Strikes Back allocated in seen_5 column was the most seen movie and the movie with the less viewers is Star Wars: Episode III Revenge of the Sith, allocated in ranking_3 column.

```python
import numpy as np
import pandas as pd 
# Split the sample between like and dislike of the Star Wars characters. To achieve this, I generate a dictionary to map
# the raw answers of the respondents with liked and disliked
mapping_dict = {'Very favorably': 'liked', np.nan: 'disliked',
               'Somewhat favorably': 'liked',
               "Neither favorably nor unfavorably (neutral)": 'disliked',
               'Unfamiliar (N/A)': 'disliked',
               'Somewhat unfavorably': 'disliked',
               'Very unfavorable': 'disliked',
               }

for col in star_wars.columns[15:29]:
    star_wars[col] = star_wars[col].map(mapping_dict)
    
```
